```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
import matplotlib as mpl
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['font.serif'] = 'Source Sans Pro Regular'
mpl.rcParams['text.usetex'] = True 
#mpl.rcParams['font.serif'] = 'Quando-Regular'
```

# The Ci and Si functions

```python
x = np.linspace(0.001,20,1000)
s = [Si(i) for i in x]
c = [Ci(i) for i in x]
```

```python
plt.figure()
plt.plot(x,s,lw=4,alpha=0.5)
plt.grid(alpha=0.7)
plt.title('The sine integral')
plt.xlabel('$x$')
plt.ylabel('$Si(x)$')
plt.savefig('Si.pdf')

plt.figure()
plt.plot(x,c,lw=4,alpha=0.5)
plt.grid(alpha=0.7)
plt.title('The cosine integral')
plt.xlabel('$x$')
plt.ylabel('$Ci(x)$')
plt.savefig('Ci.pdf')
```
